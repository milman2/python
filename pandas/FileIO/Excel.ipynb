```python
from IPython.display import Image, display
import numpy as np
import pandas as pd

display(Image(url='https://www.python.org/static/img/python-logo.png'))
```

```python
# pip install openpyxl
pd.read_excel('data/example.xlsx')
```

```python
# 시트이름으로 불러오기
pd.read_excel('data/example.xlsx', sheet_name='2nd Sheet')
```

```python
# 번호로 불러오기 (시작이 0)
pd.read_excel('data/example.xlsx', sheet_name = 0)
```

```python
# header : 열 이름(헤더)으로 사용할 행 지정 / 첫 행이 헤더가 아닌 경우 header = None

pd.read_excel('data/example.xlsx', header = 1)
```

```python
# names : 열 이름을 지정. 불어오는 열의 개수와 일치해야 한다.
pd.read_excel('data/example.xlsx', names = ['col1', 'col2'])
```

```python
# usecols : 불러올 열 지정
pd.read_excel('data/example.xlsx', usecols = ['Name', 'Grade'])
```

```python
# usecols : 번호로 지정
pd.read_excel('data/example.xlsx', usecols = [1, 2])
```

```python
# na_values : 결측치로 사용할 값 지정
pd.read_excel('data/example.xlsx', sheet_name=2, na_values = ['-', 'NA'])
```

```python
# nrows : 불러올 행 개수 제한 / 처음 ~ n 번째 행까지 불러오기
pd.read_excel('data/example.xlsx', nrows=2)
```

```python
# skiprows : 불러올 때 건너뛸 행 지정
pd.read_excel('data/example.xlsx', skiprows=2)
```

```python
# skipfooter : 불러올 때 건너뛸 행 지정 (마지막 n 행)
pd.read_excel('data/example.xlsx', skipfooter=2)
```
